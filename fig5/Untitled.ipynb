```python
import numpy as np
import matplotlib.pyplot as plt    #pylab
from numpy.fft import fft, ifft
import matplotlib as mpl
import seaborn as sns
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['text.usetex'] = True  # for using Times New Roman fonts
plt.rc('text.latex', preamble=r'\usepackage{amsmath}')  #to import latex package
#mpl.style.use('default')  #for 'seaborn' style or 'default'
```

```python
#loading data
DA15=np.loadtxt('A_sig15.txt')
DA2=np.loadtxt('A_sig2.txt')
DA25=np.loadtxt('A_sig25.txt')
DA3=np.loadtxt('A_sig3.txt')

DP15=np.loadtxt('p_sig15.txt')
DP2=np.loadtxt('p_sig2.txt')
DP25=np.loadtxt('p_sig25.txt')
DP3=np.loadtxt('p_sig3.txt')

beta=DA15[:,0]

A15=DA15[:,1]
A2=DA2[:,1]
A25=DA25[:,1]
A3=DA3[:,1]

P15=DP15[:,1]
P2=DP2[:,1]
P25=DP25[:,1]
P3=DP3[:,1]
```

```python
plt.figure()
plt.plot(beta,A15, '-o')
plt.plot(beta,A2, '-o')
plt.plot(beta,A25, '-o')
plt.plot(beta,A3, '-o')

plt.figure()
plt.plot(beta,P15, '-o')
plt.plot(beta,P2, '-o')
plt.plot(beta,P25, '-o')
plt.plot(beta,P3, '-o')
```
