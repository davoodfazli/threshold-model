```python
import numpy as np
import matplotlib.pyplot as plt    #pylab
from numpy.fft import fft, ifft
import matplotlib as mpl
import seaborn as sns
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['text.usetex'] = True  # for using Times New Roman fonts
plt.rc('text.latex', preamble=r'\usepackage{amsmath}')  #to import latex package
#mpl.style.use('default')  #for 'seaborn' style or 'default'
```

```python
#loading data 
Da1=np.loadtxt('act1.txt')
Aa=Da1[:,1]
Dp1=np.loadtxt('pro1.txt')
Pa=Dp1[:,1]

Da2=np.loadtxt('act2.txt')
Ab=Da2[:,1]
Dp2=np.loadtxt('pro2.txt')
Pb=Dp2[:,1]

Da3=np.loadtxt('act3.txt')
Ac=Da3[:,1]
Dp3=np.loadtxt('pro3.txt')
Pc=Dp3[:,1]

Da4=np.loadtxt('act4.txt')
Ad=Da4[:,1]
Dp4=np.loadtxt('pro4.txt')
Pd=Dp4[:,1]

Da5=np.loadtxt('act5.txt')
Ae=Da5[:,1]
Dp5=np.loadtxt('pro5.txt')
Pe=Dp5[:,1]

Da6=np.loadtxt('act6.txt')
Af=Da6[:,1]
Dp6=np.loadtxt('pro6.txt')
Pf=Dp6[:,1]

Da7=np.loadtxt('act7.txt')
An=Da7[:,1]
Dp7=np.loadtxt('pro7.txt')
Pn=Dp7[:,1]
```

```python
ta = Da1[:,0]
tb = Da2[:,0]
tc = Da3[:,0]
td = Da4[:,0]
te = Da5[:,0]
tf = Da6[:,0]
```

```python
from matplotlib.gridspec import GridSpec
fig=plt.figure(figsize=(15,6),facecolor='white')
gs=GridSpec(6,16)


ms=4
line=1.5
Acol='indianred'
Pcol='steelblue'


#-------------- row 1
ax1=fig.add_subplot(gs[0:2,0:6])
plt.setp( ax1.get_xticklabels(), visible=False)
ax1.plot(ta,Aa,'o-', color=Acol, lw=line, markersize=ms)
ax1.plot(ta,Pa,'*-', color=Pcol, lw=line, markersize=ms+1)
plt.ylim([-0.03,1])
plt.xlim([0,50])
plt.grid(color = 'white', linestyle = ':', linewidth = 2)
plt.title(r'$\alpha = 2$',fontsize=17)
plt.yticks(fontsize=15)


#-------------- row 1
ax2=fig.add_subplot(gs[0:2,6:])
#plt.setp( ax2.get_xticklabels(), visible=False)
plt.setp( ax2.get_yticklabels(), visible=False)
ax2.plot(tb,Ab,'o-', color=Acol, lw=line, markersize=ms)
ax2.plot(tb,Pb,'*-', color=Pcol, lw=line, markersize=ms+1)
plt.ylim([-0.03,1])
plt.xlim([0,30])
plt.xticks(fontsize=15)
plt.grid(color = 'white', linestyle = ':', linewidth = 2)
plt.title(r'$\alpha = 30$',fontsize=17)

#-------------- row 2
ax21=fig.add_subplot(gs[2:4,0:6])
plt.setp( ax21.get_xticklabels(), visible=False)
ax21.plot(ta,An,'o-', color=Acol, lw=line, markersize=ms)
ax21.plot(ta,Pn,'*-', color=Pcol, lw=line, markersize=ms+1)
plt.ylim([-0.03,1])
plt.xlim([0,30])
plt.grid(color = 'white', linestyle = ':', linewidth = 2)
#plt.title('(a)',fontsize=18)
plt.yticks(fontsize=15)



#-------------- row 2
ax3=fig.add_subplot(gs[2:4,6:11])
#plt.setp( ax3.get_xticklabels(), visible=False)
plt.setp( ax3.get_yticklabels(), visible=False)
ax3.plot(tc,Ac,'o-', color=Acol, lw=line, markersize=ms)
ax3.plot(tc,Pc,'*-', color=Pcol, lw=line, markersize=ms+1)
plt.ylim([-0.03,1])
plt.xlim([0,30])
plt.xticks(fontsize=15)
plt.grid(color = 'white', linestyle = ':', linewidth = 2)
plt.yticks(fontsize=15)



#-------------- row 2
ax4=fig.add_subplot(gs[2:4,11:])
#plt.setp( ax4.get_xticklabels(), visible=False)
plt.setp( ax4.get_yticklabels(), visible=False)
ax4.plot(td,Ad,'o-', color=Acol, lw=line, markersize=ms, label=r'$\rho_{A}$')
ax4.plot(td,Pd,'*-', color=Pcol, lw=line, markersize=ms+1, label=r'$\rho_{P}$')
#plt.legend(ncol=2 ,fontsize=16, loc ="upper right")
plt.ylim([-0.03,1])
plt.xlim([0,30])
plt.xticks(fontsize=15)
plt.grid(color = 'white', linestyle = ':', linewidth = 2)




#-------------- row 3
ax5=fig.add_subplot(gs[4:,0:6])
#plt.setp( ax1.get_xticklabels(), visible=False)
ax5.plot(te,Ae,'o-', color=Acol, lw=line, markersize=ms)
ax5.plot(te,Pe,'*-', color=Pcol, lw=line, markersize=ms+1)
plt.ylim([-0.03,1])
plt.xlim([0,30])
plt.grid(color = 'white', linestyle = ':', linewidth = 2)
plt.xlabel('$t$', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)



#-------------- row 3
ax6=fig.add_subplot(gs[4:,6:])
plt.setp( ax6.get_yticklabels(), visible=False)
ax6.plot(tf,Af,'o-', color=Acol, lw=line, markersize=ms)
ax6.plot(tf,Pf,'*-', color=Pcol, lw=line, markersize=ms+1)
plt.ylim([-0.03,1])
plt.xlim([-0.03,30])
plt.grid(color = 'white', linestyle = ':', linewidth = 2)
plt.xlabel('$t$', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)



plt.figtext(0.39, 0.81, '(a)', fontsize=16)
plt.figtext(0.445, 0.89, '(d)', fontsize=16)
plt.figtext(0.39, 0.52, '(b)', fontsize=16)
plt.figtext(0.445, 0.59, '(e)', fontsize=16)
plt.figtext(0.71, 0.59, '(f)', fontsize=16)
plt.figtext(0.39, 0.2, '(c)', fontsize=16)
plt.figtext(0.445, 0.29, '(g)', fontsize=16)


plt.figtext(0.02, 0.82, '$\sigma=1.2$', fontsize=17)
plt.figtext(0.02, 0.52, '$\sigma=1.5$', fontsize=17)
plt.figtext(0.02, 0.22, '$\sigma=4$', fontsize=17)


lgnd=fig.legend(bbox_to_anchor=(0.95,.33), ncol=1 ,fontsize=20, frameon=False)
#lgnd.legendHandles[0]._legmarker.set_markersize(8)
#lgnd.legendHandles[1]._legmarker.set_markersize(10)
plt.subplots_adjust(top=0.94, bottom=0.1, left=0.12, right=0.95, hspace=.8, wspace=.5)


plt.savefig('varVerTime.pdf')

#plt.savefig('varVerTime.jpeg')
```
