```python
import numpy as np
import matplotlib.pyplot as plt    #pylab
from numpy.fft import fft, ifft
import matplotlib as mpl
import seaborn as sns
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['text.usetex'] = True  # for using Times New Roman fonts
plt.rc('text.latex', preamble=r'\usepackage{amsmath}')  #to import latex package
#mpl.style.use('default')  #for 'seaborn' style or 'default'
```

```python
#loading data
Da12=np.loadtxt('A-sig12.txt')
A12=Da12[:,1]


Dp12=np.loadtxt('p-sig12.txt')
P12_1=Dp12[:,1]
P12_2=Dp12[:,2]

Da15=np.loadtxt('A-sig15.txt')
A15_1=Da15[:,1]
A15_2=Da15[:,2]

Da152=np.loadtxt('A-sig15_2.txt')
A152_1=Da152[:,1]
A152_2=Da152[:,2]

Dp15=np.loadtxt('p-sig15.txt')
P15_1=Dp15[:,1]
P15_2=Dp15[:,2]

Dp152=np.loadtxt('p-sig15_2.txt')
P152_1=Dp152[:,1]
P152_2=Dp152[:,2]
P152_3=Dp152[:,3]


Da4=np.loadtxt('A-sig4.txt')
A4=Da4[:,1]

Dp4=np.loadtxt('p-sig4.txt')
P4=Dp4[:,1]
```

```python
alpha12 = np.arange(0.,46.)
alpha12_1 = Dp12[:,0]
alpha15 = np.arange(0.,70.)
alpha15_1 = Dp15[:,0]
alpha15_2 = Dp152[:,0]
alpha16 = np.arange(0.,70.)
alpha16_1 = Da15[:,0]
alpha16_2 = Da152[:,0]
alpha4 = Da4[:,0]
alpha5 = Dp4[:,0]
```
