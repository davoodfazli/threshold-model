```python
import numpy as np
import matplotlib.pylab as plt
from numpy.fft import fft, ifft
#import seaborn as sns
import matplotlib as mpl

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['text.usetex'] = True  # for using Times New Roman fonts
plt.rc('text.latex', preamble=r'\usepackage{amsmath,dsfont}')  #to import latex package
#mpl.style.use('default')  #for 'seaborn' style or 'default'
```

```python
#loading data

A=np.loadtxt('A.txt')
P=np.loadtxt('P.txt')

#A=AA[0:50,:]
```

```python
# Plot

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9,4))

ax1= plt.subplot(1,2 ,1)
im=ax1.imshow(A, cmap='Reds', vmin=0, vmax=1, interpolation = "gaussian", extent=[0,1,0,1], origin='lower') #gist_heat_r
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.title(r'$\rho_{A}$', fontsize=24)
plt.ylabel(r'$\alpha$', fontsize=24)
plt.xlabel(r'$\beta$', fontsize=24)
plt.colorbar(im, fraction=0.046, pad=0.04)
ax1.set_yticks([0, .25, .5, .75, 1])
ax1.set_yticklabels([0, 25, 50, 75, 100])
ax1.set_xticks([0, .25, .5, .75, 1])
ax1.set_xticklabels([0, .25, .5, .75, 1])



ax2= plt.subplot(1,2 ,2)
im=ax2.imshow(P, cmap='Blues', vmin=0, vmax=1, interpolation = "gaussian", extent=[0,1,0,1], origin='lower')  #YlGnBu
plt.title(r'$\rho_{P}$', fontsize=24)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel(r'$\beta$', fontsize=24)
plt.colorbar(im, fraction=0.046, pad=0.04)
ax2.set_yticks([0, .25, .5, .75, 1])
ax2.set_yticklabels([0, 25, 50, 75, 100])
ax2.set_xticks([0, .25, .5, .75, 1])
ax2.set_xticklabels([0, .25, .5, .75, 1])


#fig.subplots_adjust(right=0.8)
#cbar_ax = fig.add_axes([0.9, 0.3, 0.02, 0.5])
#fig.colorbar(im, cax=cbar_ax)


plt.subplots_adjust(top=0.98, bottom=0.04, left=0.09, right=0.93, hspace=.1 ,wspace=.4)
plt.savefig('AlphaBeta.pdf')
```
